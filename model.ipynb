```python
import pandas as pd
from sklearn.linear_model import SGDRegressor
```

```python
df_train = pd.read_csv('train_pp.csv')
df_test = pd.read_csv('test_pp.csv')

model = SGDRegressor(random_state=1,learning_rate='adaptive',loss = 'huber',max_iter=10000,early_stopping=True)
x_train = df_train[['UNDER_CONSTRUCTION','RERA','BHK_NO.','SQUARE_FT','READY_TO_MOVE','RESALE']]
y_train = df_train["TARGET(PRICE_IN_LACS)"]

df_train.head()
```

```python
x_test = df_test[['UNDER_CONSTRUCTION','RERA','BHK_NO.','SQUARE_FT','READY_TO_MOVE','RESALE']]
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
df_test["TARGET(PRICE_IN_LACS)"] = y_pred
df_test.head()
```
